#  Calculate the Maximum Absolute Proportion Difference ($\text{Max } \Delta \pi$)

```r
# Rename and prepare the input data for calculation
dtu_prioritization_data <- SIG_proportions.m_subset %>%
    dplyr::rename(
        proportion = DRIMSeq_modelled_proportions,
        feature_id = transcript_id
    ) %>%
    # Ensure factors are in the correct comparison order
    mutate(
        treat = factor(treat, levels = c("LM411", "LM511")),
        time_point = factor(time_point, levels = c("t0", "t1", "t2", "t3", "t4"))
    ) %>%
    # Select only necessary columns
    dplyr::select(gene_id, gene_name, feature_id, proportion, treat, time_point, adj_pvalue)


# Calculate the Max Absolute Proportion Difference (Max Δπ)
DTU_prioritization_results <- dtu_prioritization_data %>%
    # Step 1: Pivot data to have LM411 and LM511 proportions in separate columns
    pivot_wider(
        id_cols = c(gene_id, gene_name, feature_id, time_point, adj_pvalue),
        names_from = treat,
        values_from = proportion
    ) %>%
    # Step 2: Calculate the absolute difference in proportion (pi_LM511 - pi_LM411)
    mutate(
        Delta_Proportion = LM511 - LM411,
        Abs_Delta_Proportion = abs(Delta_Proportion)
    ) %>%
    # Step 3: Group by gene and time point to find the maximum difference
    group_by(gene_id, gene_name, time_point, adj_pvalue) %>%
    summarise(
        # The transcript responsible for the largest shift
        Max_Shift_Transcript = feature_id[which.max(Abs_Delta_Proportion)],
        # The maximum absolute difference for the gene at this time point
        Max_Abs_Delta_pi = max(Abs_Delta_Proportion, na.rm = TRUE),
        # The actual signed shift for the max transcript (LM511 - LM411)
        Max_Delta_pi_Signed = Delta_Proportion[which.max(Abs_Delta_Proportion)],
        .groups = 'drop'
    ) %>%
    # Step 4: Group by gene only to find the largest shift *across all time points*
    group_by(gene_id, gene_name) %>%
    summarise(
        Overall_Max_Delta_pi = max(Max_Abs_Delta_pi, na.rm = TRUE),
        Time_Point_of_Max_Shift = time_point[which.max(Max_Abs_Delta_pi)],
        # Using indexing [1] to safely retrieve the adjusted p-value
        Adj_Pvalue = adj_pvalue[1], 
        .groups = 'drop'
    ) %>%
    # Step 5: Arrange results by the magnitude of the shift (descending)
    arrange(desc(Overall_Max_Delta_pi), Adj_Pvalue)

# Display the top 10 prioritized genes
print("Top 10 DTU Genes Prioritized by Shift Magnitude (Max Δπ):")
print(head(DTU_prioritization_results, 10))
```

## Merge with the transcript data

```r
# --- 1. Load/Prepare Data (Assuming the corrected prioritization table is final) ---

# DTU_prioritization_results_final is the gene-level summary table (from previous step)
# SIG_proportions.m_subset is the transcript-level proportions table

# Clean and rename the proportions data slightly before merging
proportions_data_clean <- SIG_proportions.m_subset %>%
    dplyr::rename(
        proportion = DRIMSeq_modelled_proportions,
        feature_id = transcript_id
    ) %>%
    # Ensure all gene identifiers are clean (though they should be already)
    mutate(gene_name = as.character(gene_name)) %>%
    dplyr::select(
        gene_id, gene_name, feature_id, 
        proportion, treat, time_point, 
        raw_pvalue = pvalue, # Keep raw p-value for reference
        adj_pvalue # Keep the gene-level FDR
    )

# Clean the prioritization table for merging
prioritization_to_merge <- DTU_prioritization_results %>%
    # Select only the new key metrics for the merge
    dplyr::select(
        gene_id, gene_name, 
        Overall_Max_Delta_pi, 
        Time_Point_of_Max_Shift
    )

# --- 2. Perform the Merge ---

# Perform a left join to attach the prioritization metrics to every transcript row
# The join keys are gene_id and gene_name
Transcript_DTU_Analysis_Table <- proportions_data_clean %>%
    left_join(prioritization_to_merge, by = c("gene_id", "gene_name")) %>%
    
    # Sort the final table by the prioritization metric
    arrange(desc(Overall_Max_Delta_pi), time_point, treat, feature_id)

# --- 3. Review the Final Merged Table ---

cat("Structure of the final merged transcript table:\n")
str(Transcript_DTU_Analysis_Table)
```

## Subset for the time points in which the time shift is max

```r
# Assuming Transcript_DTU_Analysis_Table is the merged table from the previous step.
# If you are using the originally merged table (without the final column names), 
# replace the column names accordingly.

# The Time_Point_of_Max_Shift value is repeated for every row of a gene, 
# so we can filter based on that value.

# Ensure the columns used for comparison are character or factor
Transcript_DTU_Analysis_Table <- Transcript_DTU_Analysis_Table %>%
  mutate(
    time_point = as.character(time_point),
    Time_Point_of_Max_Shift = as.character(Time_Point_of_Max_Shift)
  )

# Subset the data frame
Max_Shift_TimePoint_Subset <- Transcript_DTU_Analysis_Table %>%
  # Filter to keep only the rows where the current time_point matches 
  # the gene's Time_Point_of_Max_Shift
  filter(time_point == Time_Point_of_Max_Shift) %>%
  
  # Clean up the sorting and select final columns for review
  arrange(desc(Overall_Max_Delta_pi), time_point, treat, feature_id) %>%
  
  # Convert time_point back to ordered factor for final use if needed
  mutate(
    time_point = factor(time_point, levels = c("t0", "t1", "t2", "t3", "t4"), ordered = TRUE),
    Time_Point_of_Max_Shift = factor(Time_Point_of_Max_Shift, levels = c("t0", "t1", "t2", "t3", "t4"), ordered = TRUE)
  )

colnames(Max_Shift_TimePoint_Subset)[which(colnames(Max_Shift_TimePoint_Subset) == 'feature_id')]<-'transcript_id'
# Review the structure and the top rows of the subsetted data
cat("Structure of the Max Shift Time Point Subset:\n")
str(Max_Shift_TimePoint_Subset)

```

# Calculation of Differential Proportion ($\Delta \pi$)

```r
# Assuming Max_Shift_TimePoint_Subset is the filtered table containing only the 
# time point of max shift for each gene.

# --- 1. Calculate Delta Pi (Δπ) in a temporary table ---
# This temporary table will be a wide format, but only for the necessary columns.
Delta_pi_Temp <- Max_Shift_TimePoint_Subset %>%
    
    # Select key identifiers and the proportion/treatment columns
    dplyr::select(gene_id, gene_name, transcript_id, treat, proportion) %>%
    
    # Pivot to align LM411 and LM511 proportions
    pivot_wider(
        names_from = treat,
        values_from = proportion
    ) %>%
    
    # Calculate Delta pi (LM511 - LM411)
    mutate(
        Delta_pi = LM511 - LM411
    ) %>%
    
    # Select only the identifiers and the new Delta_pi column
    dplyr::select(gene_id, transcript_id, Delta_pi)


# --- 2. Merge Delta Pi back into the original long format ---

# Merge the Delta_pi column back into the Max_Shift_TimePoint_Subset
Max_Shift_TimePoint_with_Delta <- Max_Shift_TimePoint_Subset %>%
    
    # Perform a left join on the gene and transcript identifiers
    left_join(Delta_pi_Temp, by = c("gene_id", "transcript_id")) %>%
    
    # Re-sort for better visualization and grouping
    arrange(desc(Overall_Max_Delta_pi), time_point, treat, transcript_id)

# Review the structure and the top rows to ensure the Delta_pi column is added
cat("Structure of the final data frame with Delta_pi:\n")
str(Max_Shift_TimePoint_with_Delta)
```

# Identifying Functional Isoform Switches

```r

# Use the final annotated data frame
df_annotated <- Max_Shift_TimePoint_with_Delta_annotated

# --- 1. Identify the Most Expressed Transcript (MET) and its Biotype per Condition ---
MET_by_condition <- df_annotated %>%
    
    # Group by gene, time point, and treatment
    group_by(gene_id, gene_name, Time_Point_of_Max_Shift, treat) %>%
    
    # Find the row with the maximum proportion (the MET)
    slice_max(order_by = proportion, n = 1, with_ties = FALSE) %>% 
    ungroup() %>%
    
    # Select the key MET information
    dplyr::select(
        gene_id, gene_name, Time_Point_of_Max_Shift, treat,
        MET_ID = transcript_id,
        MET_Biotype = transcript_biotype
    )


# --- 2. Pivot to Compare METs Side-by-Side and Filter ---
Functional_Switch_Summary_DF <- MET_by_condition %>%
    
    # Pivot to align LM411 and LM511 METs side-by-side
    pivot_wider(
        id_cols = c(gene_id, gene_name, Time_Point_of_Max_Shift),
        names_from = treat,
        values_from = c(MET_ID, MET_Biotype)
    ) %>%
    
    # Check for switches
    mutate(
        ID_Switched = (MET_ID_LM411 != MET_ID_LM511),
        Biotype_Switched = (MET_Biotype_LM411 != MET_Biotype_LM511)
    ) %>%
    
    # Filter to keep ONLY genes where BOTH the ID and Biotype switched
    filter(ID_Switched == TRUE, Biotype_Switched == TRUE) %>%
    
    # Join the overall Max Delta Pi metric back for final prioritization
    left_join(
        df_annotated %>% 
            dplyr::select(gene_id, Overall_Max_Delta_pi) %>% 
            distinct(), 
        by = "gene_id"
    ) %>%
    
    # --- 3. Final Ordering and Selection ---
    
    # Sort by the magnitude of the shift
    arrange(desc(Overall_Max_Delta_pi)) %>%
    
    # Reorder and rename columns for a clean presentation
    dplyr::select(
        Gene_Symbol = gene_name,
        Gene_ID = gene_id,
        Max_Delta_pi = Overall_Max_Delta_pi,
        Max_Shift_TimePoint = Time_Point_of_Max_Shift,
        
        # Side-by-side comparison columns:
        MET_ID_LM411,
        MET_Biotype_LM411,
        MET_ID_LM511,
        MET_Biotype_LM511,
        
        # Confirmation flags
        ID_Switched,
        Biotype_Switched
    ) %>%
    
    # 4. EXPLICIT CONVERSION TO BASE R DATA FRAME
    as.data.frame()

# Return the final data frame
print("Data Frame of Genes with Functional Switches (MET ID & Biotype Changed):")
cat(str(Functional_Switch_Summary_DF))

```

# Strategy different 3'UTR

```r
# --- Script to filter the DTU list down to only 3' UTR switch genes ---
# A 3' UTR switch is defined as a gene where:
# 1. The Most Expressed Transcript (MET) ID changes between conditions.
# 2. The Biotype of the METs remains the same.
# 3. The 3' UTR length of the two METs is different.
# 4. The Coding Sequence (CDS) of the two METs is identical.

# Prerequisites: 
# max_shift_long_data (reloaded from Max_Shift_Annotated_Transcripts.tsv)
# gtf_subset_rep (The annotated GTF data frame provided in the context)

# Use the final annotated data frame
if (!exists("max_shift_long_data")) {
    stop("Error: 'max_shift_long_data' data frame (reloaded from Max_Shift_Annotated_Transcripts.tsv) is required but not found.")
}
if (!exists("gtf_subset_rep")) {
    stop("Error: 'gtf_subset_rep' (GTF annotations) data frame is required but not found.")
}

# Alias the data frame to the local variable 'df_annotated' for cleaner code
df_annotated <- max_shift_long_data

# --- A. Calculate 3' UTR Lengths from GTF ---
# Sums the length of all segments marked as 'three_prime_utr' for each transcript.
utr3_lengths <- gtf_subset_rep %>%
    # Filter for the specific feature type
    dplyr::filter(type == "three_prime_utr") %>%
    
    # Calculate the length of each UTR segment: (end - start + 1)
    dplyr::mutate(utr3_segment_length = end - start + 1) %>%
    
    # Group by transcript and sum segment lengths for total 3' UTR length
    dplyr::group_by(transcript_id) %>%
    dplyr::summarise(
        UTR3_Length = sum(utr3_segment_length, na.rm = TRUE),
        .groups = 'drop'
    )

# --- B. Calculate CDS Signature from GTF ---
# Creates a unique identifier for the CDS (coding sequence) of each transcript
# by concatenating the sorted start/end coordinates of all CDS segments.
cds_signatures <- gtf_subset_rep %>%
    dplyr::filter(type == "CDS") %>%
    
    # Sort coordinates to ensure a consistent signature regardless of row order
    dplyr::arrange(transcript_id, start) %>%
    
    # Concatenate all start and end points into a single, unique string per transcript
    dplyr::group_by(transcript_id) %>%
    dplyr::summarise(
        CDS_Signature = paste(start, end, collapse = "_"),
        .groups = 'drop'
    )


# --- 1. Identify the Most Expressed Transcript (MET), Biotype, UTR Length, and CDS Signature per Condition ---
MET_by_condition <- df_annotated %>%
    
    # Group by gene, time point, and treatment
    dplyr::group_by(gene_id, gene_name, Time_Point_of_Max_Shift, treat) %>%
    
    # Find the row with the maximum proportion (the MET)
    dplyr::slice_max(order_by = proportion, n = 1, with_ties = FALSE) %>% 
    dplyr::ungroup() %>%
    
    # Select the key MET information
    dplyr::select(
        gene_id, gene_name, Time_Point_of_Max_Shift, treat,
        MET_ID = transcript_id,
        MET_Biotype = transcript_biotype
    ) %>%
    
    # Join with the pre-calculated 3' UTR lengths
    dplyr::left_join(utr3_lengths, by = c("MET_ID" = "transcript_id")) %>%
    dplyr::mutate(UTR3_Length = tidyr::replace_na(UTR3_Length, 0)) %>%

    # NEW: Join with the pre-calculated CDS signature
    dplyr::left_join(cds_signatures, by = c("MET_ID" = "transcript_id")) %>%
    # Assign a unique placeholder for non-coding transcripts (no CDS feature)
    dplyr::mutate(CDS_Signature = tidyr::replace_na(CDS_Signature, "NON_CODING_MET"))


# --- 2. Pivot to Compare METs Side-by-Side and Filter ---
UTR3_Switch_Summary_DF <- MET_by_condition %>%
    
    # Pivot to align LM411 and LM511 METs side-by-side
    tidyr::pivot_wider(
        id_cols = c(gene_id, gene_name, Time_Point_of_Max_Shift),
        names_from = treat,
        values_from = c(MET_ID, MET_Biotype, UTR3_Length, CDS_Signature)
    ) %>%
    
    # Check for switches
    dplyr::mutate(
        # 1. MET ID must be different (Required: TRUE)
        ID_Switched = (MET_ID_LM411 != MET_ID_LM511),
        
        # 2. MET biotype must be the same (Required: FALSE)
        Biotype_Switched = (MET_Biotype_LM411 != MET_Biotype_LM511),
        
        # 3. 3' UTR length must be different (Required: TRUE)
        UTR3_Switched = (UTR3_Length_LM411 != UTR3_Length_LM511),
        
        # 4. CDS signature must be the same (Required: TRUE)
        CDS_Same = (CDS_Signature_LM411 == CDS_Signature_LM511)
    ) %>%
    
    # --- Filter to keep ONLY genes meeting the four criteria ---
    dplyr::filter(
        ID_Switched == TRUE,
        Biotype_Switched == FALSE,
        UTR3_Switched == TRUE,
        CDS_Same == TRUE
    ) %>%
    
    # Join the overall Max Delta Pi metric back for final prioritization
    dplyr::left_join(
        df_annotated %>% 
            dplyr::select(gene_id, Overall_Max_Delta_pi, adj_pvalue) %>% 
            dplyr::distinct(), 
        by = "gene_id"
    ) %>%
    
    # --- 3. Final Ordering and Selection ---
    
    # Sort by the magnitude of the shift
    dplyr::arrange(dplyr::desc(Overall_Max_Delta_pi)) %>%
    
    # Reorder and rename columns for a clean presentation
    dplyr::select(
        Gene_Symbol = gene_name,
        Gene_ID = gene_id,
        Max_Delta_pi = Overall_Max_Delta_pi,
        Adj_P_Value = adj_pvalue,
        Max_Shift_TimePoint = Time_Point_of_Max_Shift,
        
        # Side-by-side comparison columns:
        MET_ID_LM411,
        UTR3_Length_LM411,
        MET_ID_LM511,
        UTR3_Length_LM511,
        MET_Biotype_Common = MET_Biotype_LM411,
        
        # Confirmation flags
        ID_Switched,
        Biotype_Switched,
        UTR3_Switched,
        CDS_Same
    ) %>%
    
    # 4. Extract the gene symbols into a dedicated vector for plotting
    as.data.frame()

# Export the final list of gene names to be used by the visualization script
functional_switch_genes <- UTR3_Switch_Summary_DF$Gene_Symbol

cat("\n--- 3' UTR Alternative Polyadenylation (APA) Switch Filtering Complete ---\n")
cat(paste0("Identified ", length(functional_switch_genes), " genes where the MET ID and 3' UTR length switched, with identical CDS and biotype.\n"))

# Print the top 5 3' UTR switches for review
cat("\nTop 5 3' UTR/APA Switch Genes (by Max Delta Pi):\n")
print(UTR3_Switch_Summary_DF[1:min(5, nrow(UTR3_Switch_Summary_DF)), ])

# The object 'functional_switch_genes' is now available for the visualization script.
```

# Protein difference

```r
# This script identifies genes where the Most Expressed Transcript (MET) switches between conditions,
# resulting in a different protein product (i.e., a change in the Ensembl Protein ID).

# --- PREREQUISITES ---
library(dplyr)
library(tidyr)
library(data.table)

# NOTE: This script assumes 'max_shift_long_data' and 'gtf_subset_rep' are loaded in the environment.

# --- 1. PREPARE GTF FOR PROTEIN ID MAPPING ---
# Extract unique transcript-to-protein mappings from the GTF (for CDS features)
transcript_to_protein_map <- gtf_subset_rep %>%
    filter(type == "CDS", protein_id != "") %>%
    # Select transcript_id and rename protein_id directly within select for robustness
    dplyr::select(transcript_id, Ensembl_Protein_ID = protein_id) %>%
    distinct()


# --- 2. IDENTIFY THE MET AND ITS PROTEIN ID PER CONDITION ---
# Use the annotated data frame (Max_Shift_Annotated_Transcripts.tsv)
df_annotated <- max_shift_long_data

MET_by_condition <- df_annotated %>%
    # Group by gene and treatment at the max shift time point
    group_by(gene_id, gene_name, Time_Point_of_Max_Shift, treat) %>%
    
    # Find the row with the maximum proportion (the MET)
    slice_max(order_by = proportion, n = 1, with_ties = FALSE) %>%
    ungroup() %>%
    
    # Select the key MET information
    dplyr::select(
        gene_id, gene_name, Time_Point_of_Max_Shift, treat,
        MET_ID = transcript_id,
        MET_Biotype = transcript_biotype
    ) %>%
    
    # Join the Protein ID (crucial for filtering)
    left_join(transcript_to_protein_map, by = c("MET_ID" = "transcript_id"))


# --- 3. PIVOT AND FILTER FOR PROTEIN SWITCHES ---
CDS_Protein_Switch_Summary_DF <- MET_by_condition %>%
    
    # Pivot to align LM411 and LM511 METs side-by-side
    pivot_wider(
        id_cols = c(gene_id, gene_name, Time_Point_of_Max_Shift),
        names_from = treat,
        values_from = c(MET_ID, MET_Biotype, Ensembl_Protein_ID)
    ) %>%
    
    # Check for switches
    mutate(
        ID_Switched = (MET_ID_LM411 != MET_ID_LM511),
        Biotype_Switched = (MET_Biotype_LM411 != MET_Biotype_LM511),
        # CRITICAL FILTER: Check if the resulting protein ID is different
        Protein_ID_Switched = (Ensembl_Protein_ID_LM411 != Ensembl_Protein_ID_LM511)
    ) %>%
    
    # Filter to keep ONLY genes where the ID switched AND the Protein ID switched
    filter(ID_Switched == TRUE, Protein_ID_Switched == TRUE) %>%
    
    # Join the overall Max Delta Pi metric back for final prioritization
    left_join(
        df_annotated %>%
            dplyr::select(gene_id, Overall_Max_Delta_pi) %>%
            distinct(),
        by = "gene_id"
    ) %>%
    
    # --- 4. Final Ordering and Selection ---
    arrange(desc(Overall_Max_Delta_pi)) %>%
    dplyr::select(
        Gene_Symbol = gene_name,
        Gene_ID = gene_id,
        Max_Delta_pi = Overall_Max_Delta_pi,
        Max_Shift_TimePoint = Time_Point_of_Max_Shift,
        
        MET_ID_LM411,
        Ensembl_Protein_ID_LM411,
        MET_ID_LM511,
        Ensembl_Protein_ID_LM511,
        Protein_ID_Switched
    ) %>%
    as.data.frame()

# --- 5. Save the Filtered List (The required dependency file) ---
fwrite(CDS_Protein_Switch_Summary_DF, file = "CDS_Protein_Switch_Summary.tsv", sep = "\t")

print("Successfully generated and saved the list of CDS-Altering Protein Switches.")
print("File saved as: 'CDS_Protein_Switch_Summary.tsv'")
head(CDS_Protein_Switch_Summary_DF)
```

```r
# This script is specifically designed to query BioMart for functional annotations (Domains, PTMs, TM regions)
# ONLY for the proteins involved in a statistically significant CDS-altering MET switch.

# --- PREREQUISITES ---
# Load necessary libraries (assuming biomaRt, dplyr, and data.table are installed)
library(biomaRt)
library(dplyr)
library(data.table)

# --- STEP 1: Load the Filtered List of CDS-Switching METs ---
cat("Loading CDS switch summary to identify relevant proteins...\n")

# Check if the object is already in memory (as created by the previous filtering script)
if (exists("CDS_Protein_Switch_Summary_DF")) {
    cds_switch_summary <- CDS_Protein_Switch_Summary_DF
    cat("Using in-memory object 'CDS_Protein_Switch_Summary_DF'.\n")
} else if (file.exists("CDS_Protein_Switch_Summary.tsv")) {
    # Fallback: Load from the file if it exists
    cds_switch_summary <- as.data.frame(fread("CDS_Protein_Switch_Summary.tsv"))
    cat("Loaded 'CDS_Protein_Switch_Summary.tsv'.\n")
} else {
    stop("Error: CDS switch data not found. Neither in-memory object 'CDS_Protein_Switch_Summary_DF' nor file 'CDS_Protein_Switch_Summary.tsv' was found.")
}


# --- STEP 2: Extract all unique Protein IDs from the Switched METs ---

# Extract all unique Ensembl Protein IDs involved in the switch
all_protein_ids <- c(
    cds_switch_summary$Ensembl_Protein_ID_LM411, 
    cds_switch_summary$Ensembl_Protein_ID_LM511
) %>% unique() %>% na.omit() %>% as.character()

if (length(all_protein_ids) == 0) {
    stop("Error: No valid Ensembl Protein IDs found for the switched METs.")
}

print(paste("Found a highly focused list of", length(all_protein_ids), "unique Ensembl Protein IDs to query."))


# --- STEP 3: Set up the BioMart Connection ---
# IMPORTANT: Change 'mmusculus_gene_ensembl' to your organism if necessary!
ensembl <- useEnsembl(
    biomart = "ensembl", 
    dataset = "mmusculus_gene_ensembl", 
    version = 111
)

# --- STEP 4: Retrieve Annotations from Ensembl for ONLY the Switched Proteins ---
cat("Querying BioMart for Domains (Pfam/InterPro)...\n")
annotation_data <- getBM(
    attributes = c(
        "ensembl_peptide_id",
        "interpro_description",
        "pfam"
    ),
    filters = "ensembl_peptide_id",
    values = all_protein_ids, # <-- Uses the highly filtered list
    mart = ensembl
)

# --- STEP 5: Aggregate all Functional Features into a Single String ---

# Clean up PTM/Feature data and aggregate all features (Domains, PTMs, TM, etc.)
domain_map_for_filtering <- annotation_data %>%
    # Filter out empty or uninformative rows, focusing on InterPro and Pfam only
    filter(!is.na(interpro_description) | !is.na(pfam)) %>%
    
    # Generate a descriptive feature string for aggregation
    mutate(
        Feature_Type = case_when(
            !is.na(interpro_description) ~ paste0("IPRO:", interpro_description),
            !is.na(pfam) ~ paste0("PFAM:", pfam),
            TRUE ~ NA_character_
        )
    ) %>%
    
    # Group by protein and summarize all features into one unique, sorted string
    group_by(Protein_ID = ensembl_peptide_id) %>%
    summarise(
        Domain_Structure_String = paste(sort(unique(na.omit(Feature_Type))), collapse = "|"),
        .groups = 'drop'
    )

# --- STEP 6: Final Cleanup and Saving ---

final_domain_map <- domain_map_for_filtering %>%
    dplyr::select(
        Protein_ID, 
        Domain_Structure_String
    ) %>%
    as.data.frame()

# Save the map as the required input file
fwrite(final_domain_map, file = "Protein_Domain_Annotations.tsv", sep = "\t")

print("\nSuccessfully generated the highly focused protein domain map!")
print("File saved as: 'Protein_Domain_Annotations.tsv' (only includes proteins from switched METs).")
head(final_domain_map)
```

```r
# NOTE: This script assumes required packages (ggplot2, dplyr, tidyr, cowplot, ggtranscript, etc.) are loaded.
# Dataframes required: max_shift_long_data, gtf_subset_rep, normalized_counts_transcript_subset, 
# normalized_counts_genes_subset, DGE_result_subset, DTE_result_subset.

# ----------------------------------------------------------------------------------
# --- 1. DEFINE HELPER FUNCTION ---
create_transcript_label <- function(id, biotype, tag, ccds) {
    id <- as.character(id)
    tag <- if (is.null(tag) || length(tag) == 0 || is.na(tag)) "" else as.character(tag)
    ccds <- if (is.null(ccds) || length(ccds) == 0 || is.na(ccds)) "" else as.character(ccds)
    short_id <- substr(id, nchar(id) - 3, nchar(id))
    parts <- c(paste(short_id, biotype, sep = "\n"))
    if (tag != "") {
        parts <- c(parts, tag)
    }
    if (ccds != "") {
        parts <- c(parts, ccds) 
    }
    return(paste(parts, collapse = "\n"))
}
# ----------------------------------------------------------------------------------

# --- 2. Define Universal Plotting Constants and Setup ---
breaks.Rank <- (seq(0, 1, by = 0.25))
labels.Rank <- as.character(breaks.Rank)
AXIS_STROKE_MM <- 0.176 
LABEL_FONT_SIZE_MM <- 6 / ggplot2::.pt

if (!exists("path_graphs")) {
    path_graphs <- "Graphs"
}

# Setup output directory and gene list
all_dtu_genes <- unique(max_shift_long_data$gene_name)
all_dtu_genes <- all_dtu_genes[!is.na(all_dtu_genes) & all_dtu_genes != ""]
output_dir <- path_graphs

if (!dir.exists(output_dir)) {
    dir.create(output_dir)
    cat(paste0("Created directory: ", output_dir, "\n"))
}

# Global GTF Pre-Filtering (same as previous script)
if ('seqid' %in% colnames(gtf_subset_rep)) {
    colnames(gtf_subset_rep)[which(colnames(gtf_subset_rep) == 'seqid')]<-'seqnames'
}
gtf_subset_rep$seqnames <- as.character(gtf_subset_rep$seqnames)
canonical_chromosomes <- c(as.character(1:22), "X", "Y", "M", "MT")
gtf_subset_rep <- gtf_subset_rep %>%
    dplyr::filter(seqnames %in% canonical_chromosomes)
cat(paste0("Pre-filtering GTF data: Filtered to only canonical seqnames.\n"))

# ----------------------------------------------------------------------------------
# --- 3. DESEQ2 DATA PREPARATION (Converting WIDE -> LONG and Creating Lookups) ---
# ----------------------------------------------------------------------------------
cat("\nPreparing DESeq2 normalized counts data for plotting...\n")

# --- 3.1. Gene Counts Conversion ---
deseq_gene_counts_wide <- normalized_counts_genes_subset 
gene_sample_cols <- setdiff(colnames(deseq_gene_counts_wide), "gene_name")

deseq_gene_counts_long <- deseq_gene_counts_wide %>%
    tidyr::pivot_longer(
        cols = all_of(gene_sample_cols), 
        names_to = "sample_time", 
        values_to = "normalized_count"
    ) %>%
    dplyr::mutate(
        treat = sub("_R[0-9]+_t[0-9]+", "", sample_time), 
        time_point = sub(".*(t[0-9]+)$", "\\1", sample_time), 
        sample = sample_time
    ) %>%
    dplyr::select(gene_name, sample, normalized_count, treat, time_point)
cat(paste0("Gene counts converted to long format (", nrow(deseq_gene_counts_long), " rows).\n"))

# --- 3.2. Transcript Counts Conversion ---
deseq_transcript_counts_wide <- normalized_counts_transcript_subset 
transcript_sample_cols <- setdiff(colnames(deseq_transcript_counts_wide), "transcript_id")

deseq_transcript_counts_long <- deseq_transcript_counts_wide %>%
    tidyr::pivot_longer(
        cols = all_of(transcript_sample_cols), 
        names_to = "sample_time", 
        values_to = "normalized_count"
    ) %>%
    dplyr::mutate(
        treat = sub("_R[0-9]+_t[0-9]+", "", sample_time), 
        time_point = sub(".*(t[0-9]+)$", "\\1", sample_time), 
        sample = sample_time
    ) %>%
    dplyr::select(transcript_id, sample, normalized_count, treat, time_point)
cat(paste0("Transcript counts converted to long format (", nrow(deseq_transcript_counts_long), " rows).\n"))

# --- 3.3. Prepare DGE result lookup table (for Gene Plot Title) ---
DGE_lookup <- DGE_result_subset %>%
    dplyr::select(gene_name, minuslog10padj) %>%
    # UPDATED: Round to 1 decimal place
    dplyr::mutate(minuslog10padj_rounded = round(minuslog10padj, 1)) %>% 
    dplyr::distinct()
cat(paste0("DGE lookup table created (", nrow(DGE_lookup), " rows).\n"))

# --- 3.4. Prepare DTE transcript result lookup table (for Transcript Plot Labels) ---
DTE_lookup <- DTE_result_subset %>%
    dplyr::select(transcript_id, minuslog10padj) %>%
    # UPDATED: Round to 1 decimal place
    dplyr::mutate(minuslog10padj_rounded = round(minuslog10padj, 1)) %>%
    dplyr::distinct()
cat(paste0("DTE lookup table created (", nrow(DTE_lookup), " rows).\n"))

# ----------------------------------------------------------------------------------
# --- 4. Iteration Loop: Process, Plot, and Save Combined Figures ---
# ----------------------------------------------------------------------------------

cat("\nStarting iterative combined plot generation...\n")
plot_count <- 0

for (gene in all_dtu_genes) {
    
    # --- GENE EXCLUSION MECHANISM ---
    GENES_TO_EXCLUDE <- c('Lcorl;Gm3414','Nupl1','Gramd3') 
    if (gene %in% GENES_TO_EXCLUDE) {
        cat(paste0("Skipping Gene: ", gene, " (Manually excluded).\n"))
        next 
    }
    # --- END EXCLUSION MECHANISM ---
    
    cat(paste0("Processing Gene: ", gene, "...\n"))
    
    # --- 4.1. Data Preparation for Current Gene (Proportion Plot) ---
    gene_data <- max_shift_long_data %>% 
        filter(gene_name == gene) %>%
        mutate(treat = factor(treat, levels = c("LM411", "LM511")))
    
    gene_data_labeled <- gene_data %>%
        mutate(
            transcript_id = as.character(transcript_id),
            display_label = paste0(transcript_id, "\n(prop. = ", round(proportion, 2), ")")
        )

    max_delta <- unique(gene_data_labeled$Overall_Max_Delta_pi)
    adj_pvalue <- unique(gene_data_labeled$adj_pvalue)[1]
    
    depicted_order <- gene_data_labeled %>%
        dplyr::select(transcript_id, display_label) %>%
        distinct() %>%
        arrange(desc(transcript_id))
        
    display_label_order <- depicted_order %>% pull(display_label)
    transcript_id_order_from_prop <- unique(gsub("\n.*", "", display_label_order)) 
    
    gene_data_labeled$display_label <- factor(gene_data_labeled$display_label, 
                                             levels = display_label_order, 
                                             ordered = TRUE)
    
    
    # --- 4.2. Plot 1: Proportion Bar Plot (p) ---
    p <- ggplot(data = gene_data_labeled,
                aes(x = treat, y = proportion)) +
        
        geom_bar(stat = "identity", fill = "lightgray", colour = 'black') +
        
        geom_text(aes(label = display_label),
                  position = position_stack(vjust = 0.5),
                  size = LABEL_FONT_SIZE_MM,
                  color = "black",
                  family = "sans") +
        
        scale_y_continuous(name = paste("DRIMseq modelled proportion", sep = " "),
                           breaks = breaks.Rank,
                           labels = labels.Rank) + 
        
        labs(
            title = bquote(.(gene) * " (DTU: " * Delta * pi * "=" * .(round(max_delta, 3)) * ", Adj P=" * .(round(adj_pvalue, 3)) * ")"),
            x = "Treatment Group"
        ) +
        theme_classic() +
        
        theme(title = element_text(size = 8, color = "black", family = "sans"),
              axis.title.y = element_text(size = 8, color = "black", family = "sans"),
              axis.title.x = element_blank(),
              axis.text.y = element_text(angle = 0, size = 6, color = "black", family = "sans"),
              axis.text.x = element_text(angle = 0, size = 6, color = "black", family = "sans"),
              axis.line.x = element_line(size = AXIS_STROKE_MM),
              axis.ticks.x = element_line(size = AXIS_STROKE_MM),
              axis.ticks.y = element_line(size = AXIS_STROKE_MM),
              axis.line.y = element_line(size = AXIS_STROKE_MM),
              legend.position = "none")
              
    
    # ----------------------------------------------------------------------------------
    # --- 4.3. DESEQ2 EXPRESSION PLOTS (GENE & TRANSCRIPT) ---
    # ----------------------------------------------------------------------------------

    # Find the time point of max shift (used for filtering expression data)
    time_point_column <- "Time_Point_of_Max_Shift" 
    temp_tp_vec <- unique(gene_data[[time_point_column]])
    valid_tp <- temp_tp_vec[!is.na(temp_tp_vec) & temp_tp_vec != ""]
    
    if (length(valid_tp) == 0) {
        max_shift_tp <- "All"
    } else {
        max_shift_tp <- as.character(valid_tp[1])
    }
    
    
    # --- 4.3.A. GENE EXPRESSION PLOT (UPDATED TITLE) ---
    
    # 1. Fetch the DGE P-value for the title
    current_dge_padj <- DGE_lookup %>% 
        dplyr::filter(gene_name == gene) %>% 
        dplyr::pull(minuslog10padj_rounded)
        
    # Handle case where DGE result is missing
    dge_label <- if(length(current_dge_padj) > 0) {
        # Formatting to ensure one decimal place is displayed, even if it's .0
        paste0("-log10pval = ", format(current_dge_padj[1], nsmall=1)) 
    } else {
        "DGE N/A"
    }
    
    # 2. Filter gene expression data (same as before)
    gene_expr_data <- deseq_gene_counts_long %>%
        dplyr::filter(gene_name == gene, time_point == max_shift_tp) %>%
        dplyr::mutate(treat = factor(treat, levels = c("LM411", "LM511")))
    
    if (nrow(gene_expr_data) == 0) {
        # Placeholder if data is missing
        p_gene_expr <- ggplot() + 
            annotate("text", x=1, y=1, label=paste("Gene Expression Missing\n", dge_label), size=3) + 
            theme_classic() + labs(y="Log Norm. Count", x="")
        cat(paste0("Warning: Gene expression data missing for ", gene, " at ", max_shift_tp, ". Using placeholder.\n"))
    } else {
        p_gene_expr <- ggplot(data = gene_expr_data, 
                              aes(x = treat, y = log10(normalized_count + 1), fill = treat)) +
            geom_boxplot(alpha=0.5, outlier.shape = NA) + 
            geom_jitter(width=0.15, size=1.5, alpha=0.7) +
            labs(y=expression(log[10](paste("Gene Norm. Count + 1"))), 
                 x=paste("Treatment at", max_shift_tp),
                 title = dge_label) +
            scale_fill_manual(values = c("LM411" = "#1f78b4", "LM511" = "#e31a1c")) +
            theme_classic() + 
            theme(legend.position = "none", 
                  plot.title = element_text(size = 8, hjust = 0.5), 
                  axis.title.y=element_text(size=8), 
                  axis.title.x=element_text(size=8),
                  axis.text=element_text(size=6))
    }
    
    
    # --- 4.3.B. TRANSCRIPT EXPRESSION PLOT (UPDATED with DTE labels) ---

    transcript_expr_data <- deseq_transcript_counts_long %>%
        dplyr::filter(transcript_id %in% transcript_id_order_from_prop, time_point == max_shift_tp) %>%
        dplyr::mutate(treat = factor(treat, levels = c("LM411", "LM511")))

    transcript_expr_data$transcript_id <- factor(transcript_expr_data$transcript_id, 
                                                levels = transcript_id_order_from_prop, 
                                                ordered = TRUE)

    if (nrow(transcript_expr_data) == 0) {
         p_transcript_expr <- ggplot() + annotate("text", x=1, y=1, label="Transcript Expression Data Missing", size=3) + theme_classic() + labs(y="Log Norm. Count", x="")
         cat(paste0("Warning: Transcript expression data missing for ", gene, " at ", max_shift_tp, ". Using placeholder.\n"))
    } else {
        
        # 1. Merge DTE P-value for the label
        transcript_expr_data_labeled <- transcript_expr_data %>%
            dplyr::left_join(DTE_lookup, by = "transcript_id") %>%
            dplyr::mutate(
                dte_label = ifelse(
                    is.na(minuslog10padj_rounded), 
                    "-log10pval = N/A", 
                    # Formatting to ensure one decimal place is displayed
                    paste0("-log10pval = ", format(minuslog10padj_rounded, nsmall=1))
                )
            )

        # 2. Determine plot limits for positioning the geom_text label
        y_max <- max(log10(transcript_expr_data_labeled$normalized_count + 1), na.rm = TRUE)
        y_pos <- y_max * 0.95 # Position near the top

        # 3. Create the plot with geom_text
        p_transcript_expr <- ggplot(data = transcript_expr_data_labeled, 
                                    aes(x = treat, y = log10(normalized_count + 1), fill = treat)) +
            geom_boxplot(alpha=0.5, outlier.shape = NA) + 
            geom_jitter(width=0.15, size=1.5, alpha = 0.7) +
            
            # --- ADD DTE P-value label inside each facet ---
            geom_text(
                data = transcript_expr_data_labeled %>% 
                       dplyr::select(transcript_id, dte_label) %>% 
                       dplyr::distinct(),
                aes(x = 1.5, y = y_pos, label = dte_label),
                inherit.aes = FALSE,
                size = 2, 
                hjust = 0.5,
                vjust = 1
            ) +
            # -----------------------------------------------
            
            facet_wrap(~ transcript_id, nrow = 1, strip.position = "top") +
            labs(y=expression(log[10](paste("Tx Norm. Count + 1"))), 
                 x=paste("Treatment Group")) +
            scale_fill_manual(values = c("LM411" = "#1f78b4", "LM511" = "#e31a1c"), name = "Treatment") +
            theme_classic() + 
            theme(legend.position = "bottom", 
                  axis.title.y=element_text(size=8), 
                  axis.title.x=element_blank(), 
                  axis.text.y=element_text(size=6), 
                  axis.text.x=element_blank(), 
                  axis.ticks.x=element_blank(), 
                  strip.text = element_text(size=6),
                  strip.background = element_rect(fill = "gray95", color = "gray50"),
                  legend.text = element_text(size=7))
    }
    
    # ----------------------------------------------------------------------------------

    # --- 4.4. Transcript Structure Plot (transcript_plot) (No Change) ---

    REP_ggtranscript <- gtf_subset_rep %>% 
        dplyr::filter(gene_name == gene) %>%
        droplevels()

    if ('seqid' %in% colnames(REP_ggtranscript)) {
        colnames(REP_ggtranscript)[which(colnames(REP_ggtranscript) == 'seqid')]<-'seqnames'
    }
    REP_ggtranscript$seqnames <- as.character(REP_ggtranscript$seqnames)

    # PER-GENE SAFEGUARD FILTER 
    unique_seqnames_gene <- unique(REP_ggtranscript$seqnames)
    if (length(unique_seqnames_gene) > 1) {
        canonical_seqname_gene <- REP_ggtranscript %>%
            dplyr::group_by(seqnames) %>%
            dplyr::summarise(n = n(), .groups = 'drop') %>%
            dplyr::arrange(desc(n)) %>%
            dplyr::slice_head(n = 1) %>% 
            dplyr::pull(seqnames)
            
        REP_ggtranscript <- REP_ggtranscript %>% dplyr::filter(seqnames == canonical_seqname_gene)
        REP_ggtranscript$seqnames <- canonical_seqname_gene 
        
        cat(paste0("Warning: Gene ", gene, " seqnames coerced to: ", canonical_seqname_gene, ".\n"))
    }
    
    if(length(unique(REP_ggtranscript$seqnames)) > 1) {
        cat(paste0("CRITICAL FAILURE: Gene ", gene, " is still annotated on multiple seqnames. Skipping plot.\n"))
        next 
    }
    
    # Pre-calculate the multi-line Y-axis label
    REP_ggtranscript <- REP_ggtranscript %>%
        dplyr::rowwise() %>%
        dplyr::mutate(
            transcript_label = create_transcript_label(
                transcript_id, 
                transcript_biotype, 
                tag, 
                ccds_id
            )
        ) %>%
        dplyr::ungroup()
    
    # Create a temporary data frame to pull transcript labels in the desired 'depicted_order'
    label_lookup <- REP_ggtranscript %>% 
        dplyr::select(transcript_id, transcript_label) %>%
        dplyr::distinct() %>%
        dplyr::left_join(depicted_order, by = "transcript_id") %>%
        dplyr::filter(!is.na(display_label)) %>% 
        dplyr::arrange(display_label) %>%
        dplyr::pull(transcript_label)
        
    label_levels <- unique(as.character(label_lookup)) 
    
    REP_ggtranscript$transcript_label <- factor(REP_ggtranscript$transcript_label,
                                                 levels = label_levels,
                                                 ordered = TRUE)
    
    # TRANSCRIPT PLOTTING
    REP_exons <- REP_ggtranscript %>% dplyr::filter(type == "exon")

    if(length(unique(REP_exons$seqnames)) > 1) {
        cat(paste0("CRITICAL FAILURE (Exons): Gene ", gene, " is still annotated on multiple seqnames in the REP_exons object. Skipping plot.\n"))
        next 
    }
    
    REP_exons_rescaled <- shorten_gaps(
        REP_exons, 
        to_intron(REP_exons, "transcript_label"), 
        group_var = "transcript_label" 
    )
    
    transcript_plot <- REP_exons_rescaled %>%
        dplyr::filter(type == "exon") %>%
        ggplot(aes(
            xstart = start,
            xend = end,
            y = transcript_label 
        )) +
        scale_y_discrete() + 
        geom_range(height = 0.25,
            fill="steelblue"
        ) +
        geom_intron(
            data = REP_exons_rescaled %>% dplyr::filter(type == "intron"),
            aes(strand=strand),
            arrow = grid::arrow(ends = "last", length = grid::unit(0.25, "cm")),
            arrow.min.intron.length = 200
        )
    
    # THEME APPLICATION
    transcript_plot <- transcript_plot +
        theme_classic() +
        theme(
            axis.title.y=element_blank(),
            axis.title.x=element_blank(),
            axis.text.y=element_text(angle=0,size=8, color="black", family="sans"),
            axis.text.x=element_text(angle=0,size=8, color="black", family="sans"),
            axis.line.x = element_line(size = 0.4),
            axis.ticks = element_line(size = 0.4),
            axis.line.y = element_line(size = 0.4)
        ) +
        theme(
            legend.title = element_blank(),
            legend.text = element_text(size=6),
            legend.key.size = unit(0.25, 'cm'),
            legend.key.height = unit(0.25, 'cm'),
            legend.key.width = unit(0.25, 'cm'),
            legend.position="bottom"
        )
    
    
    # --- 4.5. Combine ALL Plots Side-by-Side ---
    # The layout is now: Proportion (1.5) | Gene Expr (1.2) | Transcript Expr (2.8) | Structure (3.0)
    combined_plot <- cowplot::plot_grid(p, p_gene_expr, p_transcript_expr, transcript_plot, 
                                     nrow = 1, 
                                     rel_widths = c(1.5, 1.2, 2.8, 3.0), 
                                     align = 'h', axis = 'l')

    # 4.5.B. Save R Plot Object (.rds)
    rds_filename <- file.path(output_dir, paste0(gene, "_Combined_Plot_Object.rds"))
    
    # ⭐ Action: Saving the 'combined_plot' object
    saveRDS(combined_plot, file = rds_filename)
    
    cat(paste0("   -> Saved plot object to: ", rds_filename, "\n"))


       
    # --- 4.6. Save Combined Figure ---
    filename <- file.path(output_dir, paste0(gene, "_DTU_Combined_Plot.svg"))
    
    ggsave(filename, 
           plot = combined_plot, 
           device = "svg", 
           width = 18, 
           height = 6, 
           units = "in")
    
    plot_count <- plot_count + 1
} 

cat(paste0("\n✅ Successfully finished processing ", plot_count, " individual DTU genes and saved combined figures to the '", output_dir, "' folder.\n"))
```
